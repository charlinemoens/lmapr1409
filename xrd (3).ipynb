# Xrd of PbF2
Dans ce notebook, nous étudions les phénomènes d'interphérences et de diffraction d'un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å sur le PbF2. On va pour se faire trouver les 3 premiers pics du diffractogramme pour ce rayonnement.

```python
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction import xrd 
from matplotlib import pyplot as plt
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("rxafJw6hy0L68C4ZGJP1j4UXfbE4nmS8" ) as m:
    s= m.get_structure_by_material_id("mp-315")
    
a = SpacegroupAnalyzer(s)
s_prim = a.get_primitive_standard_structure()

g = XRDCalculator(wavelength = 1.54060)
```

```python
pics = g.get_pattern(s_prim)
pic1_hkl=pics.hkls[0][0].get("hkl")
pic2_hkl=pics.hkls[1][0].get("hkl")
pic3_hkl=pics.hkls[2][0].get("hkl")

x_1=pics.x[0]
y_1=pics.y[0]
x_2=pics.x[1]
y_2=pics.y[1]
x_3=pics.x[2]
y_3=pics.y[2]


print("PREMIER PIC:")
print("             Angle 2θ du pic=", x_1 , "\n             Intensité du pic=" , y_1)
print("             Ses indices hkl associés sont:" , pic1_hkl)


print("DEUXIEME PIC:")
print("             Angle 2θ du pic=", x_2 , "\n             Intensité du pic=" , y_2)
print("             Ses indices hkl associés sont:" , pic2_hkl)


print("TROISIEME PIC:")
print("             Angle 2θ du pic=", x_3 , "\n             Intensité du pic=" , y_3)
print("             Ses indices hkl associés sont:" , pic3_hkl)

print("\n DIFFRACTOGRAMME")
g.show_plot(s_prim,two_theta_range=(0,50))
```
